```python
#제너레이터 - 이터레이터 만들 때 사용
#함수 내부에 yield 사용 시 해당 함수는 제너레이터 됨, 일반 함수와 달리 호출해도 함수 내부 실행X

#함수 선언
def test():
    print("함수 호출")
    yield test
#함수 호출
print("A지점 통과")
test()

print("B지점 통과")
test()

print(test())
```

```python
#next() - 제너레이터 함수 내부 실행, yield까지만 실행하며 next() 함수 리턴값으로 yield 뒤에 입력한 값(여기서는 1, 2)이 출력됨
#함수 선언
def test():
    print("A지점 통과")
    yield 1
    print("B지점 통과")
    yield 2
    print("C지점 통과")
        
#함수 호출
output = test()

print("D지점 통과")
a = next(output)
print(a)

print("E지점 통과")
b = next(output)
print(b)
```
