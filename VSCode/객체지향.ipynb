#학생을 매개변수로 받는 형태의 함수로 만들면 코드가 균형 잡히게 됨

```python
def create_student(name, kor, math, eng, sci):
    return {
        "name": name,
        "kor": kor,
        "math": math,
        "eng": eng,
        "sci": sci
    }
    
def score_sum(student):
    return student["kor"] + student["math"] + student["eng"] + student["sci"]
    
def score_avg(student):
    return score_sum(student) / 4

def student_to_string(student):
    return "{}\t{}\t{}".format(student["name"], score_sum(student), score_avg(student))
    
students = [create_student("윤인성", 87, 98, 88, 95),
            create_student("연하진", 87, 98, 88, 95),
            create_student("구지연", 87, 98, 88, 95),
            create_student("나선주", 87, 98, 88, 95),
            create_student("윤아린", 87, 98, 88, 95),
            create_student("윤명월", 87, 98, 88, 95),]

print("이름", "총점", "평균", sep = "\t")
for student in students:
    #점수의 총합과 평균
    print(student_to_string(student))
```

```python
#메소드 - 클래스가 가지고 있는 함수
#예>
class Student:
    def __init__(self, name, kor, math, eng, sci):  #self - '자기 자신'을 나타내는 딕셔너리
        self.name = name
        self.kor = kor
        self.math = math
        self.eng = eng
        self.sci = sci
        
    def score_sum(self):
        return self.kor + self.math + self.eng + self.sci
    
    def score_avg(self):
        return self.score_sum() / 4
    
    def to_string(self):
        return "{}\t{}\t{}".format(self.name, self.score_sum(), self.score_avg())
    
students = [
    Student("윤인성", 87, 98, 88, 95),
    Student("연하진", 87, 98, 88, 95),
    Student("구지연", 87, 98, 88, 95),
    Student("나선주", 87, 98, 88, 95),
    Student("윤아린", 87, 98, 88, 95),
    Student("윤명월", 87, 98, 88, 95),
]

print("이름", "총점", "평균", sep = "\t")
for student in students:
    #점수의 총합과 평균
    print(student.to_string())
```

```python
# 클래스 예시 1
import datetime

class Alarm:
    def __init__(self, high, coloring):
        self.high = high
        self.coloring = coloring

    def alarm(self):
        now = datetime.datetime.now()
        hour = now.hour
        min = now.minute 
        print(f"{hour}시 {min}분: {self.high} 크기의 알람, {self.coloring}")
        

# 객체 생성
al1 = Alarm(1, "띠로링")
al2 = Alarm(2, "일어나")
al3 = Alarm(3, "일어나란 말이야")

al1.alarm()
al2.alarm()
al3.alarm()
```

```python
# 클래스 예시 2
class Horse:
    def __init__(self, color, high):
        self.color = color
        self.high = high

    def horse(self):
        print(f"{self.color}색 말이 {self.high}등으로 완주했습니다.")
        if int(self.high) == 1:
            print(" -> 1등입니다! 축하드립니다.")
        if int(self.high) == 8:
            print(" -> 8등입니다! 다음에 더 잘할 거에요.")
        

# 객체 생성
h1 = Horse("갈", 1)
h2 = Horse("검정", 2)
h3 = Horse("흰", 3)
h4 = Horse("금", 4)
h5 = Horse("흰검얼룩", 5)
h6 = Horse("노란", 6)
h7 = Horse("미", 7)
h8 = Horse("회", 8)

h1.horse()
h2.horse()
h3.horse()
h4.horse()
h5.horse()
h6.horse()
h7.horse()
h8.horse()
```

```python
# 클래스 예시 3
import random

class Coffee_call:
    def __init__(self, num):
        self.num = num
        
    def coffee_call(self):
        menues = self.menu()
        print(f"{self.num}번째 손님, 주문하신 {menues} 나왔습니다.")
        
    def menu(self):
        menu_list = ["아메리카노", "라떼", "스무디"]
        menu_hotter = ["HOT", "COLD"]
        menu_flavor = ["딸기", "포도", "바나나", "복숭아", "자두"]
        choice = random.choice(menu_list)
        if choice == menu_list[0]:
            hotter = random.choice(menu_hotter)
            menu = f"{str(menu_list[0])} {hotter}"
        elif choice == menu_list[1]:
            hotter = random.choice(menu_hotter)
            menu = f"{str(menu_list[1])} {hotter}"
        elif choice == menu_list[2]:
            flavor = random.choice(menu_flavor)
            menu = f"{flavor} {str(menu_list[2])}"
        return menu
    
c1 = Coffee_call(1)
c2 = Coffee_call(2)
c3 = Coffee_call(3)
c4 = Coffee_call(4)

c1.coffee_call()
c2.coffee_call()
c3.coffee_call()
c4.coffee_call()
```
