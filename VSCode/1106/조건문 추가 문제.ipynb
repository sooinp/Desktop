```python
#쿠폰, 카드 기간 한정 할인 엔진 구현

#온라인 쇼핑몰에서 다양한 정책이 동시에 적용될 수 있는데, 모든 할인에 대해 각 조건에 따라 적용 우선순위와 중복 불가 규칙 있음.
```

1. 기본 등급 할인
    VIP: 20만원 이상 15%
    REG: 20만원 이상 10%
    그 외 할인X
2. 쿠폰 및 카드 할인 행사
    A: 7%
    B: 5%
-> 쿠폰 A와 카드 할인 동시에 존재할 경우 더 유리한 쪽만 선택. 쿠폰 B는 중복 할인 가능
3. 기간 한정 프로모션 - 행사 기간에 3% 할인 추가 적용(1~2 끝나고 마지막에 들어가는 할인임)

```python
구매금액 = int(input("구매금액:"))
회원등급 = input("회원 등급(VIP/REG):")
쿠폰종류 = input("쿠폰 종류(A/B/없음):")
카드행사참여 = input("카드 행사 참여(y/n):")
기간한정 = input("기간 한정 행사 여부(y/n):")
print("구매금액:", 구매금액, "\n회원등급:", 회원등급, "\n쿠폰종류:", 쿠폰종류, "\n카드행사참여:", 카드행사참여, "\n기간한정행사:", 기간한정)

def 회원(구매금액):
    if 회원등급 == 'VIP' and 구매금액 >= 200000:
        구매금액 *= 0.85
    elif 회원등급 == 'REG' and 구매금액 >= 200000:
        구매금액 *= 0.9
    else:
        pass
    return 구매금액
def 쿠폰(구매금액):
    if 쿠폰종류 == 'A':
        구매금액 *= 0.93
    elif 쿠폰종류 == 'B':
        if 카드행사참여 == 'y':
            구매금액 *= 0.9
        else:
            구매금액 *= 0.95
    else: #카드행사X
        pass
    return 구매금액
def 기간한정행사(구매금액):
    if 기간한정 == 'y':
        구매금액 *= 0.97
    return 구매금액

def 규칙(회원등급_규, 쿠폰종류_규, 카드행사참여_규, 기간한정_규):
    if 회원등급_규 == 'VIP':
        회원등급_규 = 'VIP 15%'
    elif 회원등급_규 == 'REG':
        회원등급_규 = 'REG 10%'
    if 쿠폰종류_규 == 'A':
        쿠폰종류_규 = 'A 7%'
    elif 쿠폰종류_규 == 'B':
        쿠폰종류_규 = 'B 5%'
    if 카드행사참여_규 == 'y':
        카드행사참여_규 = '카드행사 참여'
    elif 카드행사참여_규 == 'n':
        카드행사참여_규 = ''
    if 기간한정_규 == 'y':
        기간한정_규 = '기간한정 3%'
    elif 기간한정_규 == 'n':
        기간한정_규 = ''
    return f'{회원등급_규} {쿠폰종류_규} {카드행사참여_규} {기간한정_규}'
print("적용 규칙: {}".format(규칙(회원등급, 쿠폰종류, 카드행사참여, 기간한정)))

최종결제금액 = int(기간한정행사(쿠폰(회원(구매금액))))
print(f"최종 결제금액: {최종결제금액}원")
```
