```python
#3. 4중 for문 2차원 행렬 덧셈
A = [[1,2], [3,4]]
B = [[5,6], [7,8]]
C = [[0,0], [0,0]]
list_num = A, B, C
res_num = []
result1, result2, result3, result4 = 0, 0, 0, 0

for col1 in range(len(list_num)):
    for col2 in range(len(list_num[col1])):
        for col3 in range(len(list_num[col1][col2])):
            if col2 == col3:
                if col2 == 0:
                    result1 += (list_num[col1][col2][col3])
                    # print("res1 = ",result1)
                else:
                    result4 += (list_num[col1][col2][col3])
                    # print("res4 = ", result4)
            if col2 != col3:
                if col2 == 0:
                    result2 += (list_num[col1][col2][col3])
                    # print("res2 = ", result2)
                else:
                    result3 += (list_num[col1][col2][col3])
                    # print("res3 = ", result3)

res_num.append([result1, result2])
res_num.append([result3, result4])
# print(result1, result2, result3, result4)
for i in range(len(res_num)):
    if i == 1:
        print(res_num[i])
        break
    else:
        print(res_num[i])
```

```python
#무한 계산기
res = 0
while True:
    a = int(input("첫 번째 숫자를 입력(끝내기는 $)>>"))
    b = int(input("두 번째 숫자를 입력(끝내기는 $)>>"))
    ch = input("연산자 입력>> ")
    if a == '$' or b == '$' or ch == '$':
        break
    
    if ch == '+':
        print(f"{a} + {b} = {a+b}")
    elif ch == '-':
        print(f"{a} - {b} = {a-b}")
    elif ch == '*':
        print(f"{a} * {b} = {a*b}")
    elif ch == '/':
        print(f"{a} / {b} = {a/b}")
    elif ch == '//':
        print(f"{a} // {b} = {a//b}")
    elif ch == '%':
        print(f"{a} % {b} = {a%b}")
    elif ch == '**':
        print(f"{a} ** {b} = {a**b}")
    else:
        print("계산 가능한 연산자만 출력 가능함(+, -, *, /, //, %, **)")
```

```python
#랜덤 숫자 맞추기 게임 - 컴퓨터가 낸 1~20 사이 랜덤 숫자 하나를 사용자가 맞출 때까지 반복
import random
num = random.randint(1, 20)
count = 0

while True:
    try:
        su = int(input("입력>>"))
        if num == su:
            print(f"정답입니다! {num}을/를 맞추셨어요.")
            break
        else:
            if num > su:
                print("더 큰 숫자입니다.")
            elif num < su:
                print("더 작은 숫자입니다.")
            count += 1
            print(f"입력된 수>>{su}")
            continue
    except:
        print("숫자를 입력해야 합니다!!")
    
print(f"{count}번만에 정답을 맞춤")
```

```python
import random
num = random.randint(1, 20)
count = 1

while True:
    su = input("입력>>")
    if su.isdigit():
        su = int(su)
        if num == su:
            print(f"정답입니다! {num}을/를 맞추셨어요.")
            break
        else:
            if num > su:
                print("더 큰 숫자입니다.")
            elif num < su:
                print("더 작은 숫자입니다.")
            count += 1
            print(f"입력된 수>>{su}")
            continue
    else:
        print("숫자를 입력해야 합니다!!")
    
print(f"{count}번만에 정답을 맞춤")
```
