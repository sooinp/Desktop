```python
#문자열 - "" 혹은 ''로 묶은 집합
#더하기(+) 기호 사용해 연결, 곱하기(*) 기호를 통해 반복 가능

#len() - 괄호 안에 문자열 넣으면 문자열의 길이 출력
#upper(), lower() - 대문자, 소문자로 문자열 모두 변환
#swapcase() - 대->소, 소->대문자로 변환 출력
#title() - 아무렇게나 영단어 입력하면 띄어쓰기를 기준으로 단어 맨 앞 글자만 대문자화 해줌

#ss.count('공부') - 괄호 안의 문자열(예> 공부)이 등장하는 횟수를 ss 문자열에서 도출
#ss.find('공부') - 괄호 안의 문자열(예> 공부)이 ss 문자열에 등장하는 왼쪽부터 첫 번째 인덱스를 반환
#ss.rfind('공부') - 괄호 안의 문자열(예> 공부)이 ss 문자열에 등장하는 오른쪽부터 첫 번째 인덱스를 반환
#ss.index() - 문자열 안에서 왼쪽부터 특정 문자(또는 문자열)가 처음 나타나는 위치(인덱스)
#ss.rindex() - 문자열 안에서 오른쪽부터 특정 문자(또는 문자열)가 처음 나타나는 위치(인덱스)
#ss.startwith('파이썬', 10) - 문자열의 앞에서 10 번째가 괄호 안의 문자열(예>파이썬)이 맞는지 True/False 반환
#ss.endswith('파이썬') - 문자열의 뒤에서 첫 번째가 괄호 안의 문자열(예>파이썬)이 맞는지 True/False 반환

#문자열 공백 삭제/변경
#ss.strip() - 양쪽 공백 제거
#ss.rstrip() - 오른쪽 공백 제거
#ss.lstrip() - 왼쪽 공백 제거
#ss.replace("_", "+") - 특정 문자나 문자열을 앞의 것(예> _)에서 뒤의 것(예> +)으로 바꿈

#예>
ss = '열심히 파이썬 공부중~'
ss.replace('파이썬', 'Python')
```

```python
#문자열 변경 응용
ss = input("입력 문자열==> ")  #IT CookBook for Python
print("출력 문자열==> ", end='')
for i in range(len(ss)):
    if ss[i] != 'o':
        print(ss[i], end='')
    else:
        print('$', end='')
```

```python
#문자열.split(구분자) - 문자열을 특정 구분자(delimiter) 로 나눠서 리스트로 반환
#문자열.splitlines(keepends=False) - 문자열을 줄 단위로(split by newline) 나눠서 리스트로 반환 (이때 keepends = False면 줄바꿈 문자(\n, \r\n)를 제거, True면 미제거)
text = "Hello\nWorld\nPython"
print(text.splitlines())
print(text.splitlines(True))

print()
#구분자.join(리스트) - 리스트(또는 문자열 시퀀스)를 하나의 문자열로 합칠 때 사용

#예>
ss = 'Python을 열심히 공부 중'
print(ss.split())
ss = '하나:둘:셋'
print(ss.split(":"))
ss = '하나\n둘\n셋'
print(ss.splitlines())
ss = '%'
print(ss.join('파이썬'))
```

```python
#연/월/일 형식으로 문자열을 받아 10년 후 날짜 출력
ss = input("날짜(연/월/일) 입력>> ")
ssList = ss.split('/')
print("입력한 날짜의 10년 후 ==> ",end='')
print(str(int(ssList[0]) + 10) + '년 ', end='')
print(ssList[1] + '월 ', end='')
print(ssList[2] + '일')
```

```python
#map() - 함수명에 대입하기
#예>
before = ['2019', '12', '31']
after = list(map(int, before))
print(after)
```

```python
#문자열.center(전체길이, 채울문자) - 문자열 가운데 정렬, 양쪽에 지정한 문자(기본은 공백) 채우기
s = "hi"
print(s.center(10))
print(s.center(10, "-"))

print()
#문자열.ljust(전체길이, 채울문자) - 문자열 왼쪽 정렬, 오른쪽에 지정한 문자 채움
print(s.ljust(10))
print(s.ljust(10, "."))

print()
#문자열.rjust(전체길이, 채울문자) - 문자열 오른쪽 정렬, 왼쪽에 지정한 문자 채움 
print(s.rjust(10))
print(s.rjust(10, "*"))

print()
#문자열.zfill(전체길이) - 문자열의 왼쪽을 숫자 0으로 채워서 전체 길이를 맞추는 용도
num = "42"
print(num.zfill(5))
print("-42".zfill(5))
```
