1. append() -> aa.append(i) 형태로 리스트 맨 뒤에 항목 추가함
2. pop() -> aa.pop() 하면 리스트 맨 뒤의 항목이 삭제되며, 그 값이 남음
3. sort() -> aa.sort() 해서 리스트 항목 정렬
4. reverse() -> aa.reverse() 하면 리스트 항목 역순으로 만듦
5. index() -> aa.index(num) 형태로 괄호 속 값 num을 리스트에서 찾아 인덱스 i를 반환함
6. insert() -> aa.insert(i, value) 형태로 지정된 인덱스 i 자리에 value 값을 삽입함
7. remove() -> aa.remove(num) 형태로 괄호 속 값 num을 리스트에서 찾아 삭제함. 단, 해당 num 값이 여러 개 있는 리스트의 경우 맨 앞의 값만 없어짐
8. extend() -> aa.extend(list_a) 해서 aa 리스트 뒤에 list_a 리스트 추가하기(더하기 연산과 동일)
9. count() -> aa.count(num) 하면 리스트에서 num 값의 개수를 세줌
10. clear() -> aa.clear() 하면 리스트 비워짐. aa = [] 상태
11. del() -> del(aa[i]) 하면 aa 리스트의 i 번째 칸의 값을 삭제함
12. len() -> len(aa) 하면 aa 리스트의 요소 개수 세줌
13. copy() -> bb = aa.copy() 형태로 bb 리스트에 aa 리스트 내용을 복사
14. sorted() -> bb = sorted(aa) 형태로 aa 리스트의 항목을 정렬해서 bb 리스트에 대입함
15. set() -> set(nums) 형태로 nums 리스트 넣으면 그 리스트의 중복이 제거됨
16. zip() -> 변수1, 변수2 ~ zip(변수1용 리스트1, 변수2용 리스트2) 형태로 리스트를 한꺼번에 이용할 수 있도록 함. 대신 리스트 갯수와 변수 갯수는 맞춰줘야 함

```python
# 1) 필터링
filtered = []
for name, met in models:
    if met["f1"] >= 0.90 and met["params_m"] <= 150:
        filtered.append((name, met))  # 튜플 유지

print("1차 필터링 결과 : ", filtered)

# 2) 정렬 (f1 desc, params asc)
# key에서 튜플 반환: (-f1, params_m) → f1은 내림차순을 위해 음수화
filtered.sort(key=lambda x: (-x[1]["f1"], x[1]["params_m"]))

# 3) 랭킹 리스트(튜플)로 변환
leaderboard = []
rank = 1
for name, met in filtered:
    leaderboard.append((rank, name, met["f1"], met["params_m"]))
    rank += 1

print(leaderboard)
```
