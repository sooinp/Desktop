```python
money = int(input("교환할 돈은 얼마? "))
c500 = money // 500
print(f"500원짜리 동전==> {c500}개")
money %= 500
c100 = money // 100
print(f"100원짜리 동전==> {c100}개")
money %= 100
c50 = money // 50
print(f"50원짜리 동전==> {c50}개")
money %= 50
c10 = money // 10
print(f"10원짜리 동전==> {c10}개")
money %= 10
print("바꾸지 못한 잔돈==> {}원".format(money))
```

```python
#관계연산자 - 크거나 작음을 비교하는 것(참은 true, 거짓은 false)
#  -> 종류: ==, !=, <, >, <=, >=

#논리연산자 - and, or, not 세 가지를 이용해 관계연산자 계산에 사용
#예>
a = int(input(""))
print(a)
print((a > 100) and (a < 200))
print((a == 100) or (a == 200))
print(not(a < 100))
```

```python
#비트연산자 - 정수를 2진수 변환한 뒤 각 자리의 비트끼리 연산하는 연산 방법
#  -> 비트연산자 종류: &(and), |(or), ^(xor), ~(비트 부정), <<(비트 왼쪽 시프트), >>(비트 오른쪽 시프트)

#비트 부정 연산자 '~' - 비트 하나만 가지고 각 비트를 반대로 계산하는 연산 방법
a = 12345

#1의 보수 - 값 반전
print("~a:", ~a)    #-12346

#2의 보수 - 값 반전 +1
print("~a+1:", ~a+1)  #-12345

#왼쪽 시프트 연산자 - 왼쪽으로 시프트(<<)할 때마다 *2 효과, 시프트한 칸 수만큼 0 채우고, 빠져나간 비트는 사라짐
b = 10
print(b << 2)  #b에 왼쪽 2번 시프트해라 == b에 2를 2번 곱해라

#오른쪽 비트 연산자 - 오른쪽으로 시프트(>>)할 때마다 /2 효과, 시프트한 칸 수만큼 0 채우고, 빠져나간 비트는 사라짐
```

```python
#쇼핑몰 할인 계산기
r = int(input("구매금액: "))
print("구매금액: ", r)
h = input("회원 여부(맞으면 y, 아니면 n)): ")
print("회원 여부(맞으면 y, 아니면 n)):", h)
if r >= 100000 and h == 'y':
    r *= 0.8
elif r >= 100000 and h == 'n':
    r *= 0.9
elif r < 100000 and h == 'y':
    r *= 0.95
print("최종 결제 금액:", r)
```
