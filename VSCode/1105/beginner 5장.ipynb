```python
###종합 계산기 ###
import re

def menu(input_a):
    if input_a == '1':
        cul = input("*** 수식을 입력하세요: ")
        # list_n = [int(x) for x in cul if x.isdigit()]
        allowed = re.sub(r"[^0-9+\-*/().% ]", "", cul)
        allowed = allowed.strip()
        if not allowed:
            return None
        if '%' in allowed:
            allowed = re.sub(r"(\d+)%", r"(\1/100)", allowed)
        try:
            res = eval(allowed)
            # return res  
        except:
            pass
            # return None
        print(f"{cul}의 결과는 {res}입니다.")
    elif input_a == '2':
        a = int(input("*** 첫 번째 숫자를 입력하세요: "))
        b = int(input("*** 두 번째 숫자를 입력하세요: "))
        res = 0
        for i in range(a, b+1):
            if i == b+1:
                break
            res += i
        print(f"{a}+...+{b}는 {res}입니다.")
      
input_r = input("1. 입력한 수식 계산 2. 두 수 사이의 합계: ") 
menu(input_r)
```

```python
#삼항연산자 if문
#예>
jumsu = 55
res = ''
if jumsu >= 60:
    res = '합격'
else:
    res = '불합격'
print(res)

print("------------")
#삼항연산자 if문 줄임
res = '합격' if jumsu >= 60 else '불합격'
print(res)
```

```python
#if 응용
import random

numbers = []
for num in range(10):
    numbers.append(random.randrange(0, 10))
print("생성된 리스트:", numbers)

for num in range(10):
    if num not in numbers:
        print("숫자 %d은/는 리스트에 없네요." %num)
```
