```python
#2. 랜덤 비밀번호 생성기 - 영문 소문자 & 숫자 섞어서 8자리 비밀번호 랜덤하게 만들어 출력
import random
pwd = []
while True:
    s = random.randint(48, 122)
    if len(pwd) == 8:
        break
    
    if s >= 48 and s <= 57:
        s = chr(s)
        pwd.append(s)
    elif s >= 65 and s <= 90:
        s = chr(s)
        pwd.append(s)
    elif s >= 97 and s <= 122:
        s = chr(s)
        pwd.append(s)

for i in range(8):
    if i == 8:
        break
    pwd[0] += pwd[i]

print(pwd[0][1:])

    
```

```python
#4. 나만의 공부 일지 자동 생성기 - 과목이름, 공부내용요약 입력하면 자동으로 오늘 날짜가 포함된 파일 생성되어 study_logs 폴더에 저장
#입력: 공부한 과목을 입력하세요: Python
#     공부 내용 요약을 입력하세요: for 문과 if 문 복습
#출력: study_logs 폴더에 '2025-11-04_Python.txt'파일이 생성되었습니다!
import os
import datetime

now = datetime.datetime.now()
year = now.year
month = now.month
day = now.day

sub = input("공부한 과목을 입력하세요: ")
con = input("공부 내용 요약을 입력하세요: ")

os.mkdir("study_logs")
os.chdir("study_logs")

with open("{}-{}-{}_{}.txt".format(year, month, day, sub), "w") as file:
    file.write("{}".format(con))

print("study_logs 폴더에 '{}-{}-{:02d}_{}.txt' 파일이 생성되었습니다!".format(year, month, int(day), sub))
```

```python
#5. 장바구니 결제 시스템
#상품 목록 고정: {"사과": 1200, "바나나": 800, "딸기": 2500}
#사용자에게 상품명과 수량(int)을 반복 입력받고, q로 종료
#잘못된 상품명 -> 안내 후 다시 입력, 수량이 정수 아니거나 1 미만이면 어류 메세지 후 무시
#장바구니 합계 출력

상품목록 = {"사과": 1200, "바나나": 800, "딸기": 2500}
sum_price = 0
while True:
    input_name = []
    input_count = []
    name = input("상품명")   
    if name == 'q':
        print(f"총액: {sum_price}")
        break
    print(f"상품명(종료는 q): {name}")
    if not name in 상품목록:
        print("->오류: 목록에 없는 상품입니다.")
    else:
        count = int(input("수량"))
        if count < 1:
            print("->오류: 수량이 1개보다 작습니다.")
            continue
        print(f"수량: {count}")
        input_name.append(name)
        input_count.append(count)
        
        if name in 상품목록:
            sum_price += int(상품목록[name]) * count
            print(sum_price)
```
