```python
import tkinter as tk
from tkinter import messagebox
import datetime
import time

def show_popup():
    root = tk.Tk()
    root.withdraw()  # 메인 윈도우 숨기기
    messagebox.showinfo("알람", "지금은 오전 {now.hour}시 {now.minute}분입니다!")
    root.destroy()

def wait_for_alarm():
    print("알람 대기 중... (오전 10시가 되면 팝업이 뜹니다)")
    while True:
        now = datetime.datetime.now()
        # 오전 10시가 되면 알람 실행
        if now.hour == 10 and now.minute == 5:
            show_popup()
            break
        time.sleep(30)  # 30초마다 시간 체크

if __name__ == "__main__":
    wait_for_alarm()
```
